# Analyzing Used Car Listings on eBay Kleinanzeigen

In this guided project, we will be exploring dataset of used cars from eBay *Kleinanzeigen*, a classifieds section of the German eBay website.

The dataset was originally scraped and uploaded to kaggle by user orgesleka but the orignal dataset is no longer available. 


DataRequest sampled the dataset to 50,000 and dirtied the dataset a little bit to mimic real scraped dataset.

The data dictionary provided with data is as follows,


- <code style="background:grey">dateCrawled</code> - When this ad was first crawled. All field-values are taken from this date.

- <code style="background:grey">name</code> - Name of the car.

- <code style="background:grey">seller</code> - Whether the seller is private or a dealer.

- <code style="background:grey">offerType</code> - The type of listing

- <code style="background:grey">price</code> - The price on the ad to sell the car.

- <code style="background:grey">abtest</code> - Whether the listing is included in an A/B test.

- <code style="background:grey">vehicleType</code> - The vehicle Type.

- <code style="background:grey">yearOfRegistration</code> - The year in which the car was first registered.

- <code style="background:grey">gearbox</code> - The transmission type.

- <code style="background:grey">powerPS</code> - The power of the car in PS.

- <code style="background:grey">model</code> - The car model name.

- <code style="background:grey">kilometer</code> - How many kilometers the car has driven.

- <code style="background:grey">monthOfRegistration</code> - The month in which the car was first registered.

- <code style="background:grey">fuelType</code> - What type of fuel the car uses.

- <code style="background:grey">brand</code> - The brand of the car.

- <code style="background:grey">notRepairedDamage</code> - If the car has a damage which is not yet repaired.

- <code style="background:grey">dateCreated</code> - The date on which the eBay listing was created.

- <code style="background:grey">nrOfPictures</code> - The number of pictures in the ad.

- <code style="background:grey">postalCode</code> - The postal code for the location of the vehicle.

- <code style="background:grey">lastSeenOnline</code> - When the crawler saw this ad last online.
  

The aim of this project is to clean the sampled dataset and analyze the listed cars in eBay Kleinanzeigen.

### A brief description of our observation

- _autos.info()_ shows that 15 columnes are in object format and 5 columns are in integer formats.
- Some columns have null values but non of them have null values more than 20%.
- The column names are in Camelcase which doesn't conform to pythons naming convention.

### Converting column name convention from camel case to snake case

We will first convert column naming convention into snake case to improve readibilty and also rename some of our column names to be more descriptive with our datasets.


```python
autos.columns
```

```python
autos.rename(columns={
    "dateCrawled":"date_crawled"
    ,"offerType":"offer_type"
    ,"vehicleType":"vehicle_type"
    ,"yearOfRegistration ": "registration_year"
    ,"gearbox":"gear_box"
    ,"powerPS":"power_ps"
    ,"monthOfRegistration":"registration_month"
    ,"fuelType":"fuel_type"
    ,"notRepairedDamage":"unrepaired_damage"
    ,"dateCreated":"ad_created"
    ,"nrOfPictures":"nr_of_pictures"
    ,"postalCode":"postal_code"
}, inplace=True)

autos.head()

```

### Basic data exploration

```python
autos.describe(include="all")
```

- **Seller** & **offertype** : we can note _seller_ and _offer_type_ columns only have two unique values and have most frequency value of 49999.


- **yearOfRegistration** : Automative car was invented in the year of 1886, therefore listings that have been registered prior to this year is invalid and car registered after listing year is also invalid.


- **registration_month** : Month ranges from 1 (Januarary) to 12 (December), 0 is therefore invalid.


- **nr_of_pictures** : This column only have value of 0, This column will be removed.


- **Price** & **Odometer** : These two columns are stored as string values, We will convert these two column values into integer for better analysis



### Converting string values into int values for better analysis - Price & Odometer

#### Converting price column

```python
autos['price'].value_counts().head()
```

```python
autos['price'] = (autos['price']
                    .str.replace('$','',regex=False)
                    .str.replace(',','',regex=False)
                    .astype('int_'))
```

```python
autos['price'].value_counts().head()
```

#### Converting odometer column

```python
autos['odometer'].value_counts().head()
```

```python
autos['odometer'] = (autos['odometer']
                    .str.replace('km','',regex=False)
                    .str.replace(',','',regex=False)
                    .astype('int_'))
```

```python
autos.rename(columns={"odometer":"odometer_km"},inplace=True)
autos['odometer_km'].value_counts().head()
```
